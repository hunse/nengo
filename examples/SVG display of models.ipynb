This is the start of a system for displaying the model you are building within IPython.

To use the system, you need the following line (in the future, we may make this automatic when you do an `import nengo`).

```python
import nengo.display.svg
```

Now can now will get a simple display of what you are making.  

```python
import nengo
model = nengo.Model('test_display_svg_basic')
model.make_node('input', [0])
model.make_ensemble('A', nengo.LIF(100), 1)
model.make_ensemble('B', nengo.LIF(100), 1)
model.make_ensemble('C', nengo.LIF(100), 1)
model.make_ensemble('D', nengo.LIF(100), 1)
model.connect('input', 'A')
model.connect('A', 'B')
model.connect('B', 'C')
model.connect('A', 'C')
model.connect('B', 'B')
model.connect('C', 'D')

model
```

As with all IPython displays, it is automatically shown if the last command 
in a code block returns an unused value.  This is the case for `model.make_node`,
`model.make_ensemble`, `model.connect` and `model.add`.
